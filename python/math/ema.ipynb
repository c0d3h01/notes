```python
import numpy as np
import matplotlib.pyplot as plt

### EMA - regular interval ###
def ema(p, u):
    a = 0.1
    return a*p + (1 - a)*u

### Plot ###
N = 100

# initial price
p0 = 10

# volatility
v = 2
delta_prices = np.random.normal(0, v, N)

# prices
p = p0
prices = []
for dp in delta_prices:
    prices.append(p)
    p += dp

# moving averages
avgs = []
u = p0
for i in range(N):
    avgs.append(u)
    p = prices[i]
    u = ema(p, u)

plt.plot(prices)
plt.plot(avgs)
plt.show()
```
